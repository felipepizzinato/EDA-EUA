# Análise Exploratória sobre a Transição Energética

Introdução sobre a base de dados:     
                       
Este conjunto de dados fornece dados avaliados por medidores de eficiência energética de 2007 a 2012 para residências existentes (de uma a quatro unidades) no estado de Nova York.

# DADOS

- # Project County / Condado do Projeto:                                            
Localização do condado onde o projeto foi implementado.


- # Project City / Cidade do Projeto:                                               
A cidade onde o projeto foi implementado.


- # Project ZIP / Código Postal Projeto:                                        
O código postal da localização do projeto.


- # Climate Zone / Zona Climática:                                                
A zona climática onde o projeto foi realizado, que pode afetar o tipo de medidas de eficiência energética necessárias, com base no clima da região.


- # Weather Station / Estação Meteorológica:                                      
A estação meteorológica utilizada para obter dados climáticos históricos, que são essenciais para normalizar o uso de energia com base nas condições climáticas da área.


- # Weather Station-Normalization / Normalização da Estação Meteorológica:   
O método utilizado para ajustar os dados de consumo de energia, levando em consideração as flutuações climáticas. Essa normalização é necessária para garantir que as economias de energia sejam comparáveis independentemente das variações sazonais do clima.


- # Project Completion Date / Data de Conclusão do Projeto:                       
A data de término do projeto de eficiência energética realizado na residência.


- # Customer Type / Tipo de Cliente:                                              
O tipo de cliente para o qual o projeto foi realizado, como residência particular, empresa, ou outro tipo de cliente institucional.


- # Size of Home / Tamanho da Casa:                                              
O tamanho da residência onde o projeto foi implementado, medido em metros quadrados (m²) ou outra unidade.


- # Volume of Home / Volume da Casa:                                             
O volume da casa em metros cúbicos, o que pode influenciar a eficiência energética, já que casas maiores podem demandar mais energia para manter a temperatura adequada.


- # Number of Units / Número de Unidades:                                         
O número de unidades habitacionais dentro da residência (pode ser uma casa com vários apartamentos ou unidades dentro de um mesmo edifício).


- # Year Home Built / Ano de Construção da Casa:                                  
O ano em que a casa foi construída, pois as casas mais antigas podem ter menos eficiência energética devido à infraestrutura desatualizada.


- # Total Project Cost / Custo Total do Projeto:                                  
O custo total do projeto de retrofit ou de melhorias de eficiência energética realizadas.


- # Contractor Incentive / Incentivo ao Contratante:     
O valor do incentivo dado ao contratante para realizar o projeto de eficiência energética.


- # Total Incentives / Incentivos Totais:                                   
O total de incentivos financeiros recebidos para o projeto, que podem incluir incentivos governamentais ou de empresas fornecedoras de energia.


- # Amount Financed through Program / Valor Financiado pelo Programa:          
O valor financiado através de programas de incentivo, como os programas de eficiência energética com suporte do governo ou de empresas de energia.


- # Estimated Annual Electric Savings (kWh) / Economia Anual Estimada de Energia Elétrica (kWh):                  
A economia anual estimada de energia elétrica após as melhorias de eficiência energética, medida em quilowatt-horas (kWh).


- # Estimated Annual Gas Savings (MMBtu) / Economia Anual Estimada de Gás (MMBtu):  
A economia anual estimada de consumo de gás, medida em unidades térmicas britânicas (MMBtu).


- # Estimated First Year Energy Bill Savings / Economia Estimada na Conta de Energia do Primeiro Ano :                
O valor estimado de economia na conta de energia no primeiro ano após a implementação do projeto.


- # Baseline Electric (kWh) / Linha de Base Elétrica (kWh):         
O consumo de energia elétrica antes da implementação do projeto de eficiência energética, medido em kWh.


- # Baseline Gas (MMBtu) / Gás de Linha de Base (MMBtu):     
O consumo de gás antes do retrofit ou das melhorias implementadas no projeto, medido em MMBtu.


- # Reporting Electric (kWh) / Relatório de Eletricidade (kWh):      
O consumo de energia elétrica relatado após as melhorias de eficiência energética.


- # Reporting Gas (MMBtu) / Relatório de Gás (MMBtu):    
O consumo de gás após a implementação das melhorias do projeto.


- # Evaluated Annual Electric Savings (kWh) / Economia Anual Avaliada de Eletricidade (kWh):    
A economia anual de energia elétrica, conforme avaliado após a realização do projeto, em kWh.


- # Evaluated Annual Gas Savings (MMBtu) / Economia Anual Avaliada de Gás (MMBtu):    
A economia anual de gás, conforme avaliado após o retrofit, em MMBtu.



# INICIANDO A LIMPEZA DOS DADOS PARA GERAR OS GRÁFICOS

Importando o arquivo .csv

```python
from google.colab import files
import pandas as pd

uploaded = files.upload()

df = pd.read_csv("residential-existing-homes-one-to-four-units-energy-efficiency-meter-evaluated-project-data-2007-201-1.csv")

df.head()
```

# Verificar dados ausentes

```python
df.isna().sum()
```

# Substituindo as linhas vazias pelas médias das linhas

```python
# Preenchendo valores ausentes nas colunas "Size of Home", "Volume of Home" e "Year Home Built"
df['Size of Home'] = df['Size of Home'].fillna(df['Size of Home'].mean())
df['Volume of Home'] = df['Volume of Home'].fillna(df['Volume of Home'].mean())
df['Year Home Built'] = df['Year Home Built'].fillna(df['Year Home Built'].median())
# Preenchendo valores ausentes com a média em economias de energia
df['Estimated Annual Electric Savings (kWh)'] = df['Estimated Annual Electric Savings (kWh)'].fillna(df['Estimated Annual Electric Savings (kWh)'].mean())
df['Estimated Annual Gas Savings (MMBtu)'] = df['Estimated Annual Gas Savings (MMBtu)'].fillna(df['Estimated Annual Gas Savings (MMBtu)'].mean())


```

```python
# Verificando dados ausentes restantes
print(df.isna().sum())
```

# Decidir como vou preencher esses dados, se será com moda, média ou mediana.
A média é 1.03, entao vou preencher com a moda


```python
# Preenchendo a coluna "Number of Units" com a moda sem usar inplace
df['Number of Units'] = df['Number of Units'].fillna(df['Number of Units'].mode()[0])


```

```python
print(df.isna().sum())
```

# GRÁFICOS ABAIXO:

# Gráfico de barras comparando as economias de energia elétrica e gás.

```python
# Comparação entre economias de energia elétrica e gás
import matplotlib.pyplot as plt


plt.figure(figsize=(10,6))
df[['Estimated Annual Electric Savings (kWh)', 'Estimated Annual Gas Savings (MMBtu)']].sum().plot(kind='bar', color=['blue', 'orange'])
plt.title('Comparação entre Economias de Energia Elétrica e Gás')
plt.ylabel('Economia Total (kWh/MMBtu)')
plt.xticks(rotation=0)
plt.legend(['Energia Elétrica', 'Gás'])
plt.show()
```

# Gráfico de linha mostrando a economia anual de eletricidade ao longo dos anos de construção.

```python
# Agrupando os dados por ano e somando a economia de eletricidade
import seaborn as sns

data_grouped = df.groupby('Year Home Built')['Estimated Annual Electric Savings (kWh)'].sum().reset_index()

# Gráfico de linha com linha de tendência
plt.figure(figsize=(10,6))

# Gráfico de linha da economia de eletricidade ao longo dos anos
sns.lineplot(data=data_grouped, x='Year Home Built', y='Estimated Annual Electric Savings (kWh)',
             marker='o', color='green', label='Economia de Eletricidade')

# Adicionando a linha de tendência
sns.regplot(data=data_grouped, x='Year Home Built', y='Estimated Annual Electric Savings (kWh)',
            scatter=False, color='red', line_kws={"color": "red", "alpha": 0.7, "lw": 2}, label='Linha de Tendência')

# Definindo título, rótulos e legenda
plt.title('Crescimento da Economia de Energia Elétrica ao Longo dos Anos')
plt.xlabel('Ano de Construção')
plt.ylabel('Economia Total de Eletricidade (kWh)')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```

# Gráfico de dispersão entre "Tamanho da Casa" e "Economia Anual de Energia Elétrica".

```python
# Gráfico de dispersão com cores representando o tamanho da casa
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['Size of Home'], y=df['Estimated Annual Electric Savings (kWh)'],
                hue=df['Size of Home'], palette='viridis', marker='o')
plt.title('Relação entre Tamanho da Casa e Economia Anual de Energia Elétrica')
plt.xlabel('Tamanho da Casa (m²)')
plt.ylabel('Economia Anual de Eletricidade (kWh)')
plt.legend(title='Tamanho da Casa', loc='upper right')
plt.show()
```

# Gráfico de dispersão entre o "Custo Total do Projeto" e "Economia Anual de Eletricidade".

```python
# Gráfico de dispersão com tamanho de ponto representando o custo do projeto
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['Total Project Cost ($)'], y=df['Estimated Annual Electric Savings (kWh)'],
                size=df['Total Project Cost ($)'], sizes=(50, 300), color='red', legend=False, marker='o')
sns.regplot(data=df, x='Total Project Cost ($)', y='Estimated Annual Electric Savings (kWh)',
            scatter=False, color='blue', line_kws={"color": "blue", "alpha": 0.7, "lw": 2}, label='Linha de Tendência')
plt.title('Relação entre Custo do Projeto e Economia de Energia Elétrica')
plt.xlabel('Custo Total do Projeto ($)')
plt.ylabel('Economia Anual de Eletricidade (kWh)')
plt.legend(title='Custo do Projeto', loc='upper left')
plt.show()
```

# Gráfico de barras mostrando a economia de eletricidade em diferentes zonas climáticas (Valores absolutos).
Vale lembrar que esse gráfico apresenta resultados absolutos, abaixo terá outro gráfico com valores relativos com a quantidade de projetos por zona climática

```python
# Redução das emissões de CO₂ em áreas com maior uso de energia renovável
plt.figure(figsize=(10,6))
df.groupby('Climate Zone')['Estimated Annual Electric Savings (kWh)'].sum().plot(kind='bar', color='blue')
plt.title('Economia de Energia Elétrica por Zona Climática')
plt.xlabel('Zona Climática')
plt.ylabel('Economia Total de Eletricidade (kWh)')
plt.xticks(rotation=45)
plt.show()
```

# Descrição das zonas climáticas:
- Zona 4A - Mixed-Humid (Temperada úmida mista) - Clima temperado, com verões qunetes e invernos frios

- Zona 5A - Cold (Frio) - Clima frio com invernos severos

- Zona 6A - Cold (Frio) - Clima frio, com invernos ainda mais rigorosos que a Zona 5A.

# Gráfico de barras que apresenta a quantidade de projetos em cada região climática

```python
import matplotlib.pyplot as plt

# Contar o número de projetos por zona climática
zone_counts = df['Climate Zone'].value_counts()

# Criar o gráfico
plt.figure(figsize=(10, 6))
zone_counts.plot(kind='bar', color='skyblue')

# Adicionar título e rótulos
plt.title('Número de Projetos por Zona Climática', fontsize=14)
plt.xlabel('Zona Climática', fontsize=12)
plt.ylabel('Número de Projetos', fontsize=12)
plt.xticks(rotation=45)

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```

# Gráfico de barras apresentando a economia por região climática (valores relativos)

```python
import matplotlib.pyplot as plt

# Passo 1: Agrupar e somar a economia de energia por zona climática
energy_savings = df.groupby('Climate Zone')['Estimated Annual Electric Savings (kWh)'].sum()

# Passo 2: Contar o número de projetos por zona climática
zone_counts = df['Climate Zone'].value_counts()

# Passo 3: Calcular a economia de energia por projeto
energy_per_project = energy_savings / zone_counts

# Passo 4: Criar o gráfico
plt.figure(figsize=(10, 6))
energy_per_project.plot(kind='bar', color='green')

# Adicionar título e rótulos
plt.title('Economia de Energia por Projeto por Zona Climática', fontsize=14)
plt.xlabel('Zona Climática', fontsize=12)
plt.ylabel('Economia de Energia por Projeto (kWh)', fontsize=12)
plt.xticks(rotation=45)

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```
